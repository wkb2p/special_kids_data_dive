```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

data = pd.read_csv('Donations.csv', sep = ',', thousands = ',')
data.head(10)
```

```python
x = data['Date Created']
data['Donation Amount'] = data['Donation Amount'].dropna()
```

```python
data['DonationSize'] = np.where(data['Donation Amount'] <= 100, 0, 1)
data.head(10)
```

```python
# data.info()

data['Received Date'] = pd.to_datetime(data['Received Date'])
```

```python
data['year'] = data['Received Date'].apply(lambda x: x.year*100 + x.month)
data.head(10)
```

```python
df1 = data.groupby(['year'],as_index=False).agg({'DonationSize':'sum'})[['DonationSize', 'year']]
```

```python
df2 = data.groupby(['year'],as_index=False).agg({'DonationSize':'count'})[['DonationSize', 'year']]
```

```python
df1
```

```python
df2
```

```python
bueno_df = df1.merge(df2, on = 'year')
bueno_df
```

```python
bueno_df['under100'] = bueno_df['DonationSize_y'] - bueno_df['DonationSize_x']
bueno_df
```

```python
bueno_df['Over100'] = bueno_df['DonationSize_x']
bueno_df
```

```python
bueno_df['Total_Count'] = bueno_df['under100'] + bueno_df['Over100']
bueno_df
```

```python
bueno_df.plot.bar(x = 'year', y = 'Total_Count', stacked = True)
```

```python
df_2010_and_after = bueno_df.copy(deep=True)
```

```python
df_2010_and_after = df_2010_and_after.loc[df_2010_and_after['year'] >= 201000]
```

```python
df_2010_and_after.plot.bar(x = 'year', y = 'Total_Count', stacked = True)
```

```python
df_2010_and_after.plot.bar(x = 'year', y = ['under100', 'Over100'], stacked = True)
```
